##### 2d. Can you compare two related quantities?

Scatterplot of number of bedrooms and number of bathrooms, colored by logerror (larger negative are blue, larger positive are red, values closer to zero are more yellow).

```python
df['bbsum'] = df.num_bed + df.num_bath

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
ax.scatter(df['num_bed'],df['num_bath'], c=df['logerror'], cmap=plt.cm.RdYlBu)
plt.title('logerror for each combination of bedroom and bathroom count', size=16)
plt.xlabel('Number of Bedrooms')
plt.ylabel('Number of Bathrooms')
```

Mean logerror and count of each grouping of number of bedrooms and bathrooms.

```python
#df[['num_bed','num_bath']].groupby(['num_bed','num_bath']).size().reset_index(name='counts')
bed_bath_group = df[['num_bed','num_bath','logerror']].groupby(['num_bed','num_bath']).agg(['mean','count']).reset_index()
bed_bath_group
```

```python
fig = plt.figure(figsize=(16,12))
plt.scatter(bed_bath_group['num_bed'],bed_bath_group['num_bath'], 
           s=bed_bath_group['logerror'][['count']]/2, c=bed_bath_group['logerror'][['mean']]**.4, cmap=plt.cm.hot)
axes = plt.gca()
axes.set_xlim([-1,11])
axes.set_ylim([-1,7.5])
plt.title('logerror for each combination of bedroom and bathroom count', size=16)
plt.xlabel('Number of Bedrooms')
plt.ylabel('Number of Bathrooms')
plt.colorbar()
```

The above bubble plot shows the logerror for each grouping of number of bedrooms and number of bathrooms. The size of the circles indicate the frequency of each combination and the color indicates the mean logerror for each group. The darker the color, the lower the error. 3 bedroom 2 bath homes are the most common, but have a lower error than the homes with more bedrooms and bathrooms. The homes with 0 bedrooms (and either 0 or more than 1 or 2 bathrooms) are most likely incorrect and should be removed from the data).

##### 2f. Can you make a time-series plot?

```python
df.transactiondate = pd.to_datetime(df.transactiondate)
grouped = df.groupby('transactiondate')['logerror'].mean().abs().reset_index()
grouped
```

```python
ggplot(grouped, aes('transactiondate','logerror')) + geom_line() +\
    ggtitle('Absolute Value of logerror over time - all data')
```

```python
df['Day'] = df['transactiondate'].dt.day
grouped2 = df.groupby('Day')['logerror'].mean().abs().reset_index()
```

```python
ggplot(grouped2, aes('Day','logerror')) + geom_line() +\
    ggtitle('Absolute Value of logerror over time - by day of month')
```

There seems to be a distinct trend that the error is near the end of the month. Variations could be because the following shows a distinct dropoff on Saturdays, which can be a 7 day range surrounding the 25th day of the month.

```python
df['dayofweek'] = df['transactiondate'].dt.dayofweek
grouped3 = df.groupby('dayofweek')['logerror'].mean().abs().reset_index()
```

```python
ggplot(grouped3, aes('dayofweek','logerror')) + geom_line() +\
    ggtitle('Absolute Value of logerror over time - by day of week')
```

Clearly the lowest error is made on Saturdays. [Day 0 is Monday, day 6 is Sunday]

```python
df_abs_logerror = df
df_abs_logerror.logerror = df_abs_logerror.logerror.abs()
df_abs_logerror
```

Correlation matrix

```python
f, ax = plt.subplots(figsize=(10, 8))
corr = df_abs_logerror.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
```

In the above heatmap, the logerror is an absolute value. So looking at the correlations to this variable alone, the ideal would be the least amount of correlation, indicating the least amount of change in logerror for any change in the other variables.
