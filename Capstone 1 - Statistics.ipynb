```python
df.corr(method='pearson')
```

```python
cp = df.corr(method='pearson').abs()

s = cp.unstack()
so = s.sort_values(kind="quicksort", ascending=False)
s
```

Strongest correlation to logerror is living area, but it's still only less than 5%.

```python
df.corr(method='kendall')
```

```python
ck = df.corr(method='kendall').abs()

s = ck.unstack()
sok = s.sort_values(kind="quicksort")
sok['logerror'].nlargest(2)
```

Strongest correlation to logerror is still living area, but now it's just over 5%.

```python
df.corr(method='spearman')
```

```python
cs = df.corr(method='spearman').abs()

s = cs.unstack()
sos = s.sort_values(kind="quicksort")
sos['logerror'].nlargest(2)
```

Again, strongest correlation to logerror is living area, but now it is 7.5%, which is still weak.

Independent t-test for means
* logerror to number of bedrooms
* logerror to number of bathrooms
* logerror to sum of bedrooms and bathrooms
* logerror to living area
* logerror to lot area

For each of these t-tests, which will be performed as Welch's t-tests due to differing population variances, all null hypotheses are that the means are equal. P-values less than .05 will reject the null hypothesis in favor of the alternate, which is that the means are not equal.

```python
a = df.logerror.abs()
b = df.num_bed
scipy.stats.ttest_ind(a,b,equal_var=False)
```

```python
a = df.logerror.abs()
c = df.num_bath
scipy.stats.ttest_ind(a,c,equal_var=False)
```

```python
a = df.logerror.abs()
d = df.num_bed + df.num_bath
scipy.stats.ttest_ind(a,d,equal_var=False)
```

```python
a = df.logerror.abs()
e = df.area_living
scipy.stats.ttest_ind(a,e,equal_var=False)
```

```python
a = df.logerror.abs()
f = df.area_lot
scipy.stats.ttest_ind(a,f,equal_var=False)
```

Because all p-values are 0.0, it appears that this test is inadequate. Will proceed with machine learning operations to see if there is any connection between the dependent variable (logerror) and the independent variables.
